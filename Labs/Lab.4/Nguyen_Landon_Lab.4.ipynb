1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class counter:    
    def __init__(self,max):
        self.count=0
        self.max_val=max
    def add(self):
        if self.count<self.max_val: self.count+=1
        else: print('Error: Maximum Count Reached')
    def reset(self):
        self.count=0

## Test

count_0=counter(10)
print(count_0.count)
count_0.count=10
print(count_0.count)
count_0.add()
print(count_0.count)
count_0.reset()
print(count_0.count)
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self,shapes_0=dict()):
        self.shapes=dict()
        self.shape_names=list()
        for kv in shapes_0:
            shapes[k]=v
        
    def add(self,shape):
        if shape.name == "":
            shape.name = self.assign_name()
        
        self.shapes[shape.name]=shape
        self.shape_names.append(shape.name)

    def update(self, canvas):
        canvas.clear_canvas()
        self.paint(canvas)
        
    def paint(self,canvas):
        for shape_name in self.shape_names:
            self.shapes[shape_name].paint(canvas)
            
    def assign_name(self):
        base="shape"
        name = base+"_0"
        
        i=1
        while name in self.shapes:
            name = base+"_"+str(i)
            i+=1
            
        return name

    def __repr__(self):
        return 'RasterDrawing('+repr(self.shapes)+')'

    def save(self,filename):
        f=open(filename,"w")
        #f.write('tmp')
        f.write(self.__repr__())
        f.close()

c2=paint.Canvas(50,40)
rd=RasterDrawing()
print(rd)
rd.add(r1)
rd.add(s1)
rd.add(t1)
rd.paint(c2)
c2.display()

r2 = paint.rectangle(10,5,15,15)
rd.add(r2)
rd.update(c2)
c2.display()
```
